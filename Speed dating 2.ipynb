```python
# Define the columns you want to keep
df1  = ['age', 'like', 'guess_prob_liked', 'shared_interests_partner','expected_num_matches']

# Drop all other columns except those specified
df = df[df1]

# Return the first 20 rows to see the new updated table
df.head(20)
```

```python
# Check if there are any missing values
missingValues = df.isnull().sum()

# Print the missing values for each column if any
print(missingValues)
```

```python
# List of columns to check for null and replace with 0
columnsToFill = ['age', 'like', 'guess_prob_liked', 'shared_interests_partner','expected_num_matches']

# Replace null values in specified columns with 0
df[columnsToFill] = df[columnsToFill].fillna(0)

df.head(20)
```

```python
# Check if there are any missing values
missingValues = df.isnull().sum()

# Print the missing values for each column if any
print(missingValues)
```

```python
# Features for predicting the 'expected_num_matches'
features = df[['age', 'like', 'guess_prob_liked', 'shared_interests_partner']]

# Target is the 'expected_num_matches'
target = df['expected_num_matches']
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

# Train the Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train) # Train the model
score = model.score(X_train, y_train)  # Assign the training score
```

```python
# Print the training score
print("Training score: {:.3f}".format(score))

# Test the model
score = model.score(X_test, y_test)
print("Test score: {:.3f}".format(score))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error

# Calculate the MSE using cross-validation
mseScorer = make_scorer(mean_squared_error)

# 5-fold cross-validation
mseScores = cross_val_score(model, features, target, cv=5, scoring = mseScorer) 

# Calculate the average MSE across folds
averageMse = np.mean(mseScores)
print("Mean Squared Error (Cross-Validation):", averageMse)
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

# Generate the learning curve data
train_sizes, trainScores, testScores = learning_curve(
    model, X_train, y_train, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10)
)

# Calculate the mean of training and test scores
trainScores_mean = -trainScores.mean(axis=1)  
testScores_mean = -testScores.mean(axis=1)    

# Plot the learning curve without the fill for standard deviation
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, trainScores_mean, label="Training error", color="r")
plt.plot(train_sizes, testScores_mean, label="Cross-validation error", color="g")

# Add labels and title
plt.xlabel("Training Set Size") # Add x-axis labels
plt.ylabel("Mean Squared Error") # Add y-axis labels
plt.title("Learning Curve for the Model") # Add title
plt.legend(loc="upper right") # Add legend
plt.grid() # Add grid
plt.show() # Show the plot
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for LinearRegression
param_grid = {
    'fit_intercept': [True, False]
}

# Initialize GridSearchCV with the model and parameter grid
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the GridSearchCV to the scaled training data
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best parameters found:", grid_search.best_params_)
print("Best accuracy score:", -grid_search.best_score_)
```
