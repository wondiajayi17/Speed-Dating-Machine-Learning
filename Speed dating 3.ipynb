```python
# Define the columns you want to keep
df1  = ['age', 'attractive', 'intelligence', 'shared_interests_partner', 'pref_o_attractive', 'pref_o_intelligence', 'pref_o_shared_interests', 'like']

# Drop all other columns except those specified
df = df[df1]

# Return the first 20 rows to see the new updated table
df.head(20)
```

```python
# Check if there are any missing values
missingValues = df.isnull().sum()

# Print the missing values for each column if any
print(missingValues)
```

```python
# List of columns to check for null and replace with 0
columnsToFill = ['age', 'attractive', 'intelligence', 'shared_interests_partner', 'pref_o_attractive', 'pref_o_intelligence', 'pref_o_shared_interests', 'like']

# Replace null values in specified columns with 0
df[columnsToFill] = df[columnsToFill].fillna(0)

df.head(20)
```

```python
# Check if there are any missing values
missingValues = df.isnull().sum()

# Print the missing values for each column if any
print(missingValues)
```

```python
# Select features based on available columns
features = ['age', 'attractive', 'intelligence', 'shared_interests_partner','pref_o_attractive', 'pref_o_intelligence', 'pref_o_shared_interests']

# The target variable
target = 'like' 
```

```python
# Extract features (X) and target (y) from the dataset
X = df[features].values
y = df[target].values
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler() 
X_train = scaler.fit_transform(X_train) # Fit and transform the training data
X_test = scaler.transform(X_test) # Transform the test data
```

```python
from sklearn.neural_network import MLPRegressor

# Train an ANN model for interactive preference modeling
model = MLPRegressor(random_state=0, hidden_layer_sizes=(16, 16), max_iter=600)
model.fit(X_train, y_train) # Fit the model
```

```python
# Print the training score
score = model.score(X_train, y_train) 
print("Training score: {:.3f}".format(score)) 

# Print the test score
score = model.score(X_test, y_test) 
print("Test score: {:.3f}".format(score))
```

```python
from sklearn.model_selection import cross_val_score

# Perform 5-fold cross-validation
cvScores = cross_val_score(model, X, y, cv=5, scoring='r2')

# Print individual R^2 scores for each fold
print("Cross-validated R^2 scores for each fold:", cvScores)

# Calculate and print the average cross-validated R^2 score
print("Average Cross-validated R^2 score:", cvScores.mean())
```

```python
import matplotlib.pyplot as plt

# Create a bar plot
plt.figure(figsize=(10, 5))
folds = np.arange(1, len(cvScores) + 1) # Generate x-axis values for the number of folds
plt.bar(folds, cvScores, color="blue") # Create a bar plot with R^2 scores for each fold

# Label the plot
plt.xlabel("Fold") # Set the x-axis label
plt.ylabel("RÂ² Score") # Set the y-axis label
plt.title("RÂ² Score for Each Cross-Validation Fold") # Set the title of the plot
plt.xticks(folds) # Set the x-ticks to be the number of folds
plt.grid(axis="y") # Add a grid for better readability

# Display the plot
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPRegressor

# Define the parameter grid
param_grid = {
    
    'hidden_layer_sizes': [(16,), (32, 16), (64, 32), (64, 64)],  # Number of neurons in each layer
    'alpha': [0.0001, 0.001, 0.01],  # Regularization parameter values
    'learning_rate_init': [0.001, 0.01, 0.1],  # Initial learning rate values
    'max_iter': [200, 500, 1000]  # Maximum number of iterations
}

# Initialize the GridSearchCV object
grid_search = GridSearchCV(
    
    estimator=MLPRegressor(random_state=0), # Initialize the MLPRegressor model
    param_grid=param_grid, # Use the defined hyperparameter grid that was created above
    cv=5, # Use 5-fold Cross-validation fold, which averages model performance over 5 different data splits.
    scoring='r2', # Use R-squared score for evaluation
    n_jobs=-1, # Use all available CPUs for faster computation.
)

# Fit the GridSearchCV object to the data
grid_search.fit(X_train, y_train)

# Print the best parameters and the best cross-validation score
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation score: {:.3f}".format(grid_search.best_score_))

# Get the best model
bestModel = grid_search.best_estimator_
```
