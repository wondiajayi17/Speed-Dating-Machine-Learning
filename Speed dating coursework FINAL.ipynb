```python
# Use dummies 
target = pd.get_dummies(df['match'], drop_first=True)  # drop_first avoids creating an extra column

# Return the first 10 rows
target.head(10)
```

```python
# Select the relevant features and target
features = df[['age', 'age_o', 'like', 'guess_prob_liked']]  

# Define the target variable
target = df['match']
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# Create a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train) # Train the model
score = model.score(X_train, y_train)  # Assign the training score

# Print the training score
print("Training score: {:.3f}".format(score))

# Test the model
score = model.score(X_test, y_test)
print("Test score: {:.3f}".format(score)) # Print the training score
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics

# Make predictions on the test set
y_pred = model.predict(X_test)

# Generate the confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)

# Display the confusion matrix using a heatmap
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=['No Match', 'Match'])

# Plot the confusion matrix
cm_display.plot() # Display the confusion matrix    
plt.xlabel('Predicted Labels') # Label the x-axis
plt.ylabel('True Labels') # Label the y-axis
plt.title('Confusion Matrix for Match Predictions') # Title of the plot
plt.show() # Display the plot
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100], # Regularisation strength
    'penalty': ['l1', 'l2'], # Regularisation type
    'solver': ['liblinear', 'saga'] # The algorithm used to optimise the model
}

# Create a logistic regression model
model = LogisticRegression(max_iter=1000)

# Create the GridSearchCV object
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')

# Fit the grid search model
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best parameters found: ", grid_search.best_params_) # The parameters that got the highest accuracy across all folds of cross-validation.
print("Best accuracy score: {:.3f}".format(grid_search.best_score_)) # The best cross-validated accuracy score achieved with the best parameters.
```

```python
# Select features based on available columns
features = ['age', 'attractive', 'intelligence', 'shared_interests_partner','pref_o_attractive', 'pref_o_intelligence', 'pref_o_shared_interests']

# The target variable
target = 'like' 
```

```python
# Extract features (X) and target (y) from the dataset
X = df[features].values
y = df[target].values
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

# Standardise the features
scaler = StandardScaler() 
X_train = scaler.fit_transform(X_train) # Fit and transform the training data
X_test = scaler.transform(X_test) # Transform the test data
```

```python
from sklearn.neural_network import MLPRegressor

# Train an ANN model for interactive preference modeling
model = MLPRegressor(random_state=0, hidden_layer_sizes=(16, 16), max_iter=600)
model.fit(X_train, y_train) # Fit the model
```

```python
# Print the training score
score = model.score(X_train, y_train) 
print("Training score: {:.3f}".format(score)) 

# Print the test score
score = model.score(X_test, y_test) 
print("Test score: {:.3f}".format(score))
```

```python
from sklearn.model_selection import cross_val_score

# Perform 5-fold cross-validation
cvScores = cross_val_score(model, X, y, cv=5, scoring='r2')

# Print individual R^2 scores for each fold
print("Cross-validated R^2 scores for each fold:", cvScores)

# Calculate and print the average cross-validated R^2 score
print("Average Cross-validated R^2 score:", cvScores.mean())
```

```python
import matplotlib.pyplot as plt

# Create a bar plot
plt.figure(figsize=(10, 5))
folds = np.arange(1, len(cvScores) + 1) # Generate x-axis values for the number of folds
plt.bar(folds, cvScores, color="blue") # Create a bar plot with R^2 scores for each fold

# Label the plot
plt.xlabel("Fold") # Set the x-axis label
plt.ylabel("RÂ² Score") # Set the y-axis label
plt.title("RÂ² Score for Each Cross-Validation Fold") # Set the title of the plot
plt.xticks(folds) # Set the x-ticks to be the number of folds
plt.grid(axis="y") # Add a grid for better readability

# Display the plot
plt.show()
```
