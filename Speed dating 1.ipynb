```python
# Define the columns you want to keep
df1  = ['age', 'age_o', 'like', 'guess_prob_liked', 'match'] 

# Drop all other columns except those specified
df = df[df1]

# Return the first 20 rows to see the new updated table
df.head(20)
```

```python
# Check if there are any missing values
missingValues = df.isnull().sum()

# Print the missing values for each column if any
print(missingValues)
```

```python
# List of columns to check for null and replace with 0
columnsToFill = ['age', 'age_o', 'like', 'guess_prob_liked', 'match']

# Replace null values in specified columns with 0
df[columnsToFill] = df[columnsToFill].fillna(0)

df.head(20)
```

```python
# Check if there are any missing values
missingValues = df.isnull().sum()

# Print the missing values for each column if any
print(missingValues)
```

```python
# Use dummies 
target = pd.get_dummies(df['match'], drop_first=True)  # drop_first avoids creating an extra column

# Return the first 10 rows
target.head(10)
```

```python
# Select the relevant features and target
features = df[['age', 'age_o', 'like', 'guess_prob_liked']]  

# Define the target variable
target = df['match']
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# Create a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train) # Train the model
score = model.score(X_train, y_train)  # Assign the training score

# Print the training score
print("Training score: {:.3f}".format(score))

# Test the model
score = model.score(X_test, y_test)
print("Test score: {:.3f}".format(score)) # Print the training score
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics

# Make predictions on the test set
y_pred = model.predict(X_test)

# Generate the confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)

# Display the confusion matrix using a heatmap
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=['No Match', 'Match'])

# Plot the confusion matrix
cm_display.plot() # Display the confusion matrix    
plt.xlabel('Predicted Labels') # Label the x-axis
plt.ylabel('True Labels') # Label the y-axis
plt.title('Confusion Matrix for Match Predictions') # Title of the plot
plt.show() # Display the plot
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100], # Regularization strength
    'penalty': ['l1', 'l2'], # Regularization type
    'solver': ['liblinear', 'saga'] # The algorithm used to optimize the model
}

# Create a logistic regression model
model = LogisticRegression(max_iter=1000)

# Create the GridSearchCV object
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')

# Fit the grid search model
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best parameters found: ", grid_search.best_params_) # The parameters that got the highest accuracy across all folds of cross-validation.
print("Best accuracy score: {:.3f}".format(grid_search.best_score_)) # The best cross-validated accuracy score achieved with the best parameters.
```
